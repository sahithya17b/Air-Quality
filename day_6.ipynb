```python

#task1 : design a calculator using oops concepts
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, a, b):
        self.result = a + b
        return self.result

    def subtract(self, a, b):
        self.result = a - b
        return self.result

    def multiply(self, a, b):
        self.result = a * b
        return self.result

    def divide(self, a, b):
        if b != 0:
            self.result = a / b
            return self.result
        else:
            return "Error: Division by zero"

    def power(self, a, b):
        self.result = a ** b
        return self.result

    def modulus(self, a, b):
        self.result = a % b
        return self.result

calc = Calculator()
print(calc.add(10, 5))
print(calc.subtract(10, 5))
print(calc.multiply(10, 5))
print(calc.divide(10, 5))
print(calc.power(2, 3))
print(calc.modulus(10, 3))
#task2: write code for multiple, multilevel, hierarchical,hybrid
#Multiple Inheritance
class Addition:
    def add(self, a, b):
        return a + b

class Subtraction:
    def subtract(self, a, b):
        return a - b

class Calculator(Addition, Subtraction):
    pass

calc = Calculator()
print(calc.add(10, 5))
print(calc.subtract(10, 5))
#Multilevel Inheritance
class Base:
    def greet(self):
        return "Hello from Base"

class Intermediate(Base):
    def welcome(self):
        return "Welcome from Intermediate"

class Derived(Intermediate):
    def message(self):
        return "Message from Derived"

obj = Derived()
print(obj.greet())
print(obj.welcome())
print(obj.message())
class Parent:
    def show(self):
        return "Parent method"

class ChildOne(Parent):
    def child_one_method(self):
        return "ChildOne method"

class ChildTwo(Parent):
    def child_two_method(self):
        return "ChildTwo method"

c1 = ChildOne()
c2 = ChildTwo()
print(c1.show())
print(c1.child_one_method())
print(c2.show())
print(c2.child_two_method())
class A:
    def method_a(self):
        return "Method from A"

class B(A):
    def method_b(self):
        return "Method from B"

class C(A):
    def method_c(self):
        return "Method from C"

class D(B, C):
    def method_d(self):
        return "Method from D"

obj = D()
print(obj.method_a())
print(obj.method_b())
print(obj.method_c())
print(obj.method_d())
#task3: write code for over riding & overloading
#Method Overriding
class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

a = Animal()
d = Dog()
print(a.speak())
print(d.speak())
#Method Overloading
class Calculator:
    def multiply(self, a, b=None):
        if b is not None:
            return a * b
        else:
            return a * a

calc = Calculator()
print(calc.multiply(5))      # Single argument: returns square
print(calc.multiply(5, 3))   # Two arguments: returns product
```
