```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.linspace(0, 10, 100)
y = np.sin(x)
data = pd.DataFrame({'x': x, 'y': y})

sns.lineplot(data=data, x='x', y='y', color='b')
plt.title("Sine Wave (Seaborn Line Plot)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Box Plot - Total Bill by Day and Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time", s=100)
plt.title("Scatter Plot - Total Bill vs Tip")
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Violin Plot - Distribution of Total Bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.countplot(data=tips, x="day", hue="sex")
plt.title("Count of Customers by Day and Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", estimator="mean")
plt.title("Average Total Bill by Day and Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 12)
sns.heatmap(data, annot=True, fmt=".1f", cmap="coolwarm")
plt.title("Random Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex", cmap="viridis")
plt.suptitle("Joint Plot - Total Bill vs Tip", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill", fill=True, color='green')
plt.title("KDE Plot - Distribution of Total Bill")
plt.show()
```
