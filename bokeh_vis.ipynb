```python
from bokeh.plotting import figure, show
from math import pi
from bokeh.transform import cumsum
from bokeh.palettes import Category20c
import pandas as pd

data = pd.Series({'A':30, 'B':45, 'C':25})
data = data.reset_index(name='value').rename(columns={'index':'category'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[len(data)]

p = figure(height=300, title="Pie Chart", toolbar_location=None, x_range=(-0.5,1))
p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True),
        end_angle=cumsum('angle'),
        fill_color='color', legend_field='category', source=data)
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np
data = np.random.randn(200)
hist, edges = np.histogram(data, bins=10)
p = figure(title="Histogram")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="green", line_color="white")
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np
data = np.random.randn(50)
p = figure(title="Box Plot (approx)")
p.segment(1, np.percentile(data, 25), 1, np.percentile(data, 75), line_width=6, color="navy")
p.circle(1, np.median(data), size=10, color="red")
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np
from scipy.stats import gaussian_kde

data = np.random.normal(0, 1, 300)
x = np.linspace(-3, 3, 100)
kde = gaussian_kde(data)
y = kde(x) / kde(x).max() * 0.4
p = figure(title="Violin Plot (approx)")
p.patch(np.concatenate([y, -y[::-1]]), np.concatenate([x, x[::-1]]), color="purple", alpha=0.6)
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np

data = np.random.rand(5, 5)
p = figure(title="Heatmap")
p.image(image=[data], x=0, y=0, dw=5, dh=5, palette="Viridis256")
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np
x = np.linspace(-5, 5, 40)
y = np.sin(x)
p = figure(title="Wireframe Surface (2D View)")
for i in range(1, 5):
    p.line(x, y*i/5, alpha=0.6)
show(p)
```

```python
from bokeh.plotting import figure, show
import numpy as np
x, y = np.random.rand(20), np.random.rand(20)
sizes = np.random.randint(10, 50, 20)
p = figure(title="Bubble Chart")
p.circle(x, y, size=sizes, fill_color="orange", alpha=0.5)
show(p)
```
